```python
import csv
import numpy as np
import random as rd
import pylab as pl
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import linear_model as lm
from sklearn.metrics import roc_curve, auc
```

###1. Lood Data

```python
ifile = open('../data/train.csv','rb')
data  = csv.reader(ifile, delimiter=',',quotechar='|')
digits = []
target = []
images = []
skip   = True
for rows in data:
    if skip:
        skip = False
        continue
    target.append(float(rows[0]))
    digitArray = np.array(map(float, rows[1:]))
    digits.append(digitArray)
    images.append(digitArray.reshape((28,28)))
target = np.array(target)
digits = np.array(digits)
images = np.array(images)
ifile.close()

print "dim(digits) =", digits.shape
plt.figure(figsize=(20,10), dpi=600)
for i in range(10):
    plt.subplot(1,10,(i+1))
    print target[i],
    pl.imshow(images[i],cmap=pl.cm.gray_r)
pl.show()
```

###2. Preprocessing

```python
# I am mimicing R to create a function to generate a partition of the data set into training set and test set
def createDataPartition(data, times=1, p=0.6, seed=5425):
    """
    Function  :  createDataPartition(data, times=1, p=0.6)
    Purpose   :  Create partition(s) of the given dataset for training and testing
    Input     :  data  : non-empty dataset (list, array, numpy array, or pandas series or data.frame etc.)
                 times : a natural number (default value=1) indicating how many partitions to create
                 p     : a rational number in [0, 1] indicating the portion of the data for training
                 seed  : a natural number (default value=5425) defining the seed for generating sequence of random numbers
    Output    :  a tuple of the two indices for the observations partitioned into the training set and the testing set respectively
    Author    :  Chuan Zhang
    Created   :  11/25/2014
    Copyright :  (c) Chuan 2014
    """
    import random as rd
    nSamples  = np.array(data).shape[0]
    rd.seed(seed)
    inTrain   = rd.sample(xrange(nSamples), int(p*nSamples))
    inTest    = filter(lambda x: x not in set(inTrain), range(nSamples))
    return (inTrain, inTest)

idx = createDataPartition(digits)
print len(idx[0]), len(idx[1]), digits.shape

train_x = digits[idx[0]]
train_y = target[idx[0]]
test_x  = digits[idx[1]]
test_y  = target[idx[1]]
```

###3. Fit Logistic Model with One-VS-All 

```python
from sklearn import metrics
from sklearn import preprocessing

# Binarize the output
train_y = preprocessing.label_binarize(train_y, classes=range(10))
test_y = preprocessing.label_binarize(test_y, classes=range(10))
```

```python
from sklearn import multiclass as mc

# Learn to predict each class against the other
classifier = mc.OneVsRestClassifier(lm.LogisticRegression(C=1e5, tol=1e-1))
score_y = classifier.fit(train_x, train_y).decision_function(test_x)
print type(score_y), score_y.shape, type(test_y), test_y.shape
```

```python
# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()
n_classes = test_y.shape[1]
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(test_y[:, i], score_y[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
```

```python
# Compute micro-average ROC curve and ROC area
fpr["micro"], tpr["micro"], _ = roc_curve(test_y.ravel(), score_y.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])
```

```python
# Plot of a ROC curve for a specific class
plt.figure()
class_ = 8
plt.plot(fpr[class_], tpr[class_], label='ROC curve (area = %0.2f)' % roc_auc[class_])
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve of class {0}'.format(class_))
plt.legend(loc="lower right")
plt.show()
```

```python
# Plot ROC curve
plt.figure()
plt.plot(fpr["micro"], tpr["micro"],
         label='micro-average ROC curve (area = {0:0.2f})'
               ''.format(roc_auc["micro"]))
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label='ROC curve of class {0} (area = {1:0.2f})'
                                   ''.format(i, roc_auc[i]))
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
#plt.legend(loc="lower right")
plt.legend(bbox_to_anchor=(1.05, 1), loc=0, borderaxespad=0.)
plt.title('Multi-Class ROC Curves for Logistic Regression')
plt.show()
```
